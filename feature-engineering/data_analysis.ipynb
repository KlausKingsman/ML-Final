```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
malware_data_filepath = 'datasets/Obfuscated/Obfuscated-MalMem2022_edited.csv'
malware_data = pd.read_csv(malware_data_filepath)
malware = malware_data.drop(columns=['Class', 'Category'])
X = malware_data.drop(columns=[
                'Class',
                'Category',
                'svcscan.interactive_process_services', 
                'handles.nport', 
                'modules.nmodules',
                'pslist.nprocs64bit', 
                'callbacks.ngeneric'])
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(malware.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix Heatmep")
plt.show()
```

```python
# run this at your own risk
for i in range(len(malware_data.columns)):
    for j in range(i+1, len(malware_data.columns)):
        feature1 = malware_data.columns[i]
        feature2 = malware_data.columns[j]
        
        plt.figure()
        plt.scatter(malware_data[feature1], malware_data[feature2])
        plt.title(f'{feature1} vs {feature2}')
        plt.xlabel(feature1)
        plt.ylabel(feature2)
        plt.show()
```
